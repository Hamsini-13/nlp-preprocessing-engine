```python
Task 1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing, "Love" and "love" are treated as different words if case conversion is not applied. This is because NLP models consider uppercase and lowercase letters separately. This can lead to duplication and inconsistency in the dataset. By converting all text into lowercase, both words become the same, which helps in improving consistency and model performance.


2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many common words such as "is", "the", and "and". These words do not add much meaning but increase noise in the data. This can make the model less efficient, increase processing time, and reduce accuracy because important words get less attention.


3. Provide two real-world scenarios where removing stopwords can be harmful.

First, in sentiment analysis, removing words like "not" can completely change the meaning of a sentence. For example, "I am not happy" becomes "I am happy", which is incorrect.

Second, in question answering systems, words like "what", "where", and "how" are important to understand the question. Removing them can make the sentence meaningless or confusing.


4. Explain the difference between stemming and lemmatization.

Stemming is a technique that removes suffixes from words using simple rules. It is fast but may produce incorrect or incomplete words. For example, "better" may become "bet".

Lemmatization is a more advanced technique that converts words into their base or dictionary form using linguistic rules. It produces meaningful words. For example, "better" becomes "good".

In short, stemming is faster but less accurate, while lemmatization is slower but more accurate.
```

```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)

    text = text.lower()

    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r'[^a-z\s]', '', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    exceptions = {"no", "not"}
    tokens = [word for word in tokens if len(word) > 2 or word in exceptions]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)

    results.append((sentence, tokens, cleaned))
```

```python
for sentence, tokens, cleaned in results:
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = round(sum(len(w) for w in tokens)/total, 2) if total else 0

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Length:", avg_len)
    print("-" * 50)
```

```python
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(test_sentences)

print(output)
```
